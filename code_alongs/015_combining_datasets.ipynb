### concat
pandas .concat() method concatenates dataframe row-, or columnwise, 
with optional set logic (union or intersection) of the indexes on the other axis.

concat simply stacks multiple dataframes togheter either vertically, or horizonetely, after aligning on row/column.

```python
# sample data
left = sample_df("A0", "D3", prefix="L")
right = sample_df("A0", "D3")

hdisplay([left, right], ["Left", "Right"])
```

```python
hdisplay([
    pd.concat([left, right], axis="index"),
    pd.concat([left, right], axis="columns")],
    ["axis='index'", "axis='columns"]
)
```

## join
join first aligns the index of two dataframes, an then pick up the remaining columns from the aligned rows of each dataframe

the **how** can be any of the folowing:
- **left** get *all* rows from the left table, and join matching rows from the right table. **(default)**
- **right** get *all* rows from both table, and join matching rows from the left table.
- **outer** get *all* rows from the right table, and join matching rows on both table. (union)
- **inner** Get only rows that exists in both tables. (intersection)
- **cross** get every possible combination of rows from both tables. length of new table == len(left) * len(right)

```python
# new sample data
left = sample_df("A0", "D3", prefix="L").add_prefix("L")
right = sample_df("C2", "F5", prefix="R").add_prefix("R")
hdisplay([left, right], ["Left", "Right"])
```

```python
hdisplay([
    left.join(right, how="left"),
    left.join(right, how="right")],
    ["how='left'", "how='right'"]
)
```

```python
hdisplay([
    left.join(right, how="inner"),
    left.join(right, how="outer")],
    ["how='inner'", "how='outer'"]
)
```

```python
left.join(right, how="cross")
```

## Merge
pandas .merge() method is similar to .join but is more versetile,
aand allow us to specify columns beside the index to join on for both dataframes.

note : *how* parameter works th same as for .join(), but the default for merge() is "inner", not "left

```python
# new sample data
left = sample_df("A0", "E3").add_prefix("L")
left.loc[:, "LF"] = ["F10", "F11", "F12", "F13"]
right = sample_df("F10", "J13").add_prefix("R")
hdisplay([left, right], ["Left", "Right"])
```

```python
left.merge(right, how="inner", left_on='LF', right_on='RF')
```
