```python
import pandas as pd

athlete_events = pd.read_csv("../extras/athlete_events.csv")

noc_regions = pd.read_csv("../extras/noc_regions.csv")
```

```python
median_age = athlete_events['Age'].mean()
min_age = athlete_events['Age'].min()
max_age = athlete_events['Age'].max()
std_age = athlete_events['Age'].std()

print(f"Median:{median_age}")
print(f"Minimum:{min_age}")
print(f"Max:{max_age}")
print(f"std:{std_age}")
```

```python
import seaborn as sns

Sex_comparison = athlete_events['Sex']

sns.set_palette(['blue', 'pink'])

sns.histplot(Sex_comparison)
```
