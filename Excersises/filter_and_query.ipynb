### Q1: How many cars from Jaguar are in the dataset, and what is the price of the most expensive one?

### Answer
###   count=3    price=36000.0

```python
jaguar_count = autos.query("`make` == 'jaguar'").count()
print(f"count={jaguar_count['make']}")

jaguar_max_price = autos.query("`make` == 'jaguar'")["price"].max()
print(f"price={jaguar_max_price}")
```

### Q2: How many cars from Toyota are in the dataset, and what is the price of the most expensive one?

### Q3: What is the length, width and height of the most expensive car in the entire dataset?

### Answer
###   length=199.2    width=72.0    height=55.4

```python
max_price = autos.loc[autos["price"].idxmax()]
car_length = max_price["length"]
car_width = max_price["width"]
car_height = max_price["height"]
print(f"length={car_length} width={car_width} height={car_height}")
```

### Q4: What is the lowest price per horsepower in the dataset, and what brand ("make") is that car?

### Answer
###   price per horsepower=72.84    brand=Toyota

```python
autos['price_per_hp'] = autos['price'] / autos['horsepower']
lowest_price_hp = autos['price_per_hp'].min()
price_per_hp = autos['price_per_hp'].idxmin()
hp_brand = autos.loc[price_per_hp, 'make']

print(f"price per horsepower={lowest_price_hp} brand={hp_brand}")
```

### Q5: How many of the cars in the dataset has as many cylinders as they have doors?

### Answer
###   cars=95

```python
matching_parts = autos[autos["num-of-cylinders"] == autos["num-of-doors"]]
num_of_matching_cars = matching_parts.shape[0]

print(f"cars={num_of_matching_cars}")
```

### T1: Print a string to inform the user of the price difference between the cheapest and the most expensive car in the dataset.

### Solution
###   Example:
###   The price difference between the cheapest and the most expensive car is 40282.0

```python
max_car_price = autos["price"].max()
min_car_price = autos["price"].min()
difference_price = max_car_price - min_car_price

print(f"the difference between the cheapest and the most expensive car is {difference_price}")
```

### T2: Ask the user to input a brand, then print the price range for that brand.
